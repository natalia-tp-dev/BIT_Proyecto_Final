##### Observamos que existen columnas con valores como "si" o "no", se pretende convertirlas a valores booleanos, ademas columnas como el estrato se convertiran a enteras

```python
df_data_types = df_propiedades.dtypes
df_data_types
```

##### Tipificamos las columnas de [remodelado, deposito, zona_de_lavanderia, gas, parqueadero] como booleanos con el fin de reemplazar el 1 por true y el 0 por false, ademas, modificamos el nombre, incluyendo un 'incluye_', con el fin de mejorar la legibilidad

```python
def tipificarColumnas(df, columnas):
    df_modificado = df.copy()
    
    for columna in columnas:
        nueva_col = f'incluye_{columna}'
        df_modificado[nueva_col] = df_modificado[columna].map({
            1: True, 
            0: False,
            'No Tiene': False,
            'Comunal': True,
            'Si': True,
            'No': False
        }).astype(bool)
    
    df_modificado.drop(columns=columnas, inplace=True)
    
    return df_modificado

df_mod = tipificarColumnas(df_propiedades, ['remodelado', 'deposito', 'zona_de_lavanderia', 'gas', 'parqueadero'])

```

## Histogramas de precio y de logaritmo de precio

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
sns.histplot(df_mod['precio'], kde=True, bins=50, color='skyblue')
plt.title('Distribución del Precio (Escala Original)')
plt.xlabel('Precio de la Propiedad')
plt.ylabel('Frecuencia')



df_mod['log_precio'] = np.log1p(df_mod['precio']) 
plt.subplot(1, 2, 2) 
sns.histplot(df_mod['log_precio'], kde=True, bins=50, color='salmon')
plt.title('Distribución de log(Precio)')
plt.xlabel('Logaritmo del Precio de la Propiedad')
plt.ylabel('Frecuencia')

plt.tight_layout() 
plt.show()
```

##### El precio original está muy sesgado hacia la derecha, con muchos valores bajos y pocos valores extremadamente altos (distribución típica de precios).
##### Al aplicar el logaritmo, la distribución se vuelve más simétrica y cercana a una normal (Campana de Gauss), lo cual es ideal para aplicar nuestro modelo.

##### Agregamos la columna de logaritmo de precio a nuestro dataframe para trabajar con esa variable.

```python
df_mod['log_precio'] = np.log1p(df_mod['precio'])
df_mod['log_precio'].describe()
```

## Analizar cómo afectan los valores extremos a los datos
#### Para el análisis, se calculan el primer cuartil (Q1) y el tercer cuartil (Q3), y posteriormente el rango intercuartílico (IQR) para cada una de las variables, luego de eso, eliminamos los outliers y comparamos la desviacion estandar del df original con el modificado, como resultado observamos que la dispersion de los datos disminuyo en el nuevo dataset.

```python
stats_before = df_mod['precio'].describe().loc[['mean', 'std']]

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

df_clean = remove_outliers(df_mod, 'precio')

stats_after = df_clean['precio'].describe().loc[['mean', 'std']]

comparison = pd.DataFrame({
    'Antes (Media)': [stats_before['mean']],
    'Después (Media)': [stats_after['mean']],
    'Antes (Desv. Est.)': [stats_before['std']],
    'Después (Desv. Est.)': [stats_after['std']]
}, index=['precio'])

display(comparison)
```

## Relacion entre la tasa de hurtos y el precio
