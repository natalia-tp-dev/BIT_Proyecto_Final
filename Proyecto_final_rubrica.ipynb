## **1. Entregables**

Cada equipo deberá entregar dos productos:

### **A. Documento escrito (máximo 10 páginas)**
Un informe técnico que contenga:

1. **Título del Proyecto**
2. **Integrantes del equipo**
3. **Introducción y Justificación**  
   Breve descripción del contexto, problema a resolver y relevancia del proyecto.
4. **Objetivos del Proyecto**
5. **Análisis Exploratorio de Datos (EDA)**  
   Presentar solo los hallazgos más llamativos, con explicaciones claras.
6. **Procesamiento y Limpieza de Datos**  
   Describir transformaciones clave, tratamiento de nulos, codificación, normalización, etc., con su justificación.
7. **Modelado**  
   - Selección del algoritmo (y su justificación)  
   - Resultados obtenidos (métricas y visualizaciones)  
   - Discusión crítica del desempeño del modelo
8. **Discusión de Resultados**  
   Análisis de los hallazgos y su impacto.
9. **Conclusiones**  
   Reflexión general sobre lo aprendido, limitaciones y potencial futuro.

---

### **B. Presentación ejecutiva (máximo 10 minutos)**
Una exposición gerencial para **“vender” la idea o el proyecto**. Deberá incluir:

1. **Introducción y Justificación** (contexto atractivo y convincente)
2. **Objetivos del Proyecto**
3. **Hallazgos clave del EDA** (solo lo más relevante)
4. **Procesamiento de Datos** (transformaciones o limpiezas críticas)
5. **Modelado**  
   - Algoritmo seleccionado  
   - Vista minable o visualización del modelo  
   - Métricas principales (solo las más significativas)
6. **Resultados y Conclusiones**
7. **Siguientes pasos / futuro del proyecto**

---

## **2. Evaluación del Proyecto**

### Ponderación final:
- **60% Presentación oral (ejecutiva)**
- **40% Documento escrito (informe técnico)**

### Mecanismo de evaluación:
- La **presentación será evaluada por coevaluación**:  
  Cada grupo de trabajo (1 voto por grupo), así como el docente, calificarán todas las presentaciones (excepto la suya y la de su equipo).
- Se usará una escala del **1 al 5**, con criterios como:  
  _entendimiento, claridad, estructura, valor del proyecto, impacto visual_.
- La **nota final de la presentación** será el **promedio de todas las coevaluaciones recibidas**.
- La **nota del documento escrito** será asignada por el docente, según criterios de calidad técnica, profundidad y claridad.

---

## **3. Recomendaciones**

- Practiquen la presentación para ajustarse al tiempo máximo (10 minutos).
- Ubiquen su proyecto en un **contexto atractivo para una audiencia ejecutiva**.
- No saturen las diapositivas de texto; privilegien visuales, gráficas y mensajes clave.
- En el informe técnico, sean **precisos** y **justifiquen cada decisión técnica**.
